```python
import numpy as np
import matplotlib
import matplotlib.pyplot as plt 
from scipy import optimize  

from algorithms import eEuler, iEuler, RK4
```

```python
# wichtig für A4a
def function_a(t, u):
    return -20 * u

def an_method_a(t):
    return 5 * np.exp(-20 * t)

# wichtig für A4b
def DGLs(t, Y_vec):
    Y_1, Y_2 = Y_vec[0], Y_vec[1]
    dY_1_dt =  Y_2
    dY_2_dt = -Y_1
    return np.array([dY_1_dt,dY_2_dt])
```

```python
# Parameter:
# aufgabe: "a" oder "b"
# solver: "eEuler", "iEuler" oder "RK4"

def Aufgabe_main(aufgabe="a", solver="eEuler"):
    T_INTERVAL = [0, 10]
    STEPS = 500
    
    params = {
    'figure.figsize' : [8,5], 'axes.titlesize' : 14, 'axes.labelsize' : 16,
    'xtick.labelsize' : 14, 'ytick.labelsize' : 14 }

    if aufgabe == "a":
        Y0 = [5]
        if solver == "eEuler":
            t_array, Y_array = eEuler(fun=function_a, Y0=Y0, t=T_INTERVAL, steps=STEPS).call()
        if solver == "iEuler":  
            t_array, Y_array = iEuler(fun=function_a, Y0=Y0, t=T_INTERVAL, steps=STEPS).call()
        if solver == "RK4":
            t_array, Y_array = RK4(fun=function_a, Y0=Y0, t=T_INTERVAL, steps=STEPS).call()

        matplotlib.rcParams.update(params) 
        plt.title("Aufgabe a) mit " + solver)
        plt.xlabel("t")
        plt.ylabel("Y_1")
        plt.plot(t_array, Y_array[:, 0], c="blue", label="Y_1")
        plt.legend(loc='best', fontsize=16)
        plt.show()



    if aufgabe == "b":
        Y0 = [0,1]
        if solver == "eEuler":
            t_array, Y_array = eEuler(DGLs, Y0=Y0, t=T_INTERVAL, steps=STEPS).call()
        if solver == "iEuler":  
            t_array, Y_array = iEuler(DGLs, Y0=Y0, t=T_INTERVAL, steps=STEPS).call()
        if solver == "RK4":
            t_array, Y_array = RK4(DGLs, Y0=Y0, t=T_INTERVAL, steps=STEPS).call()

        matplotlib.rcParams.update(params) 
        plt.title("Aufgabe b) mit " + solver)
        plt.xlabel("t")
        plt.ylabel("Y_1, Y_2")
        plt.plot(t_array, Y_array[:, 0], c="blue", label="Y_1")
        plt.plot(t_array, Y_array[:, 1], c="red", label="Y_2")
        plt.legend(loc='best', fontsize=16)
        plt.show()
```

```python
Aufgabe_main(aufgabe="a", solver="eEuler")
```
