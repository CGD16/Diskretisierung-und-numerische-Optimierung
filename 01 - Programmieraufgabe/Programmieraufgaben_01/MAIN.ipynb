```python
import numpy as np
import matplotlib.pyplot as plt

from eEuler import eEuler
from iEuler import iEuler
from rk4 import RK4
```

```python
u0 = 5 
N = 0.2
T_INTERVAL = [0, N]
STEPS = 10000

def function_a(t, u):
    return -20 * u

def an_method_a(t):
    return 5 * np.exp(-20 * t)


def function_b(t, u):
    return np.array([[0, 1],[-1, 0]]) * u

def an_method_b(t):
    pass


eEuler_solver = eEuler(fun=function_a, t=T_INTERVAL, Y0=u0, steps=STEPS)
iEuler_solver = iEuler(fun=function_a, t=T_INTERVAL, Y0=u0, steps=STEPS)
rk4_solver = RK4(fun=function_a, t=T_INTERVAL, Y0=u0, steps=STEPS)

t_list, eEuler_solution = eEuler_solver.call()
t_list, iEuler_solution = iEuler_solver.call()
t_list, rk4_solution = rk4_solver.call()

exact_solution = np.array([an_method_a(t) for t in t_list])



# Visualisierung
plt.figure(figsize=(10, 6))
plt.plot(t_list, exact_solution, label="Analytische Lsung", linestyle="-", color="blue")
plt.plot(t_list, eEuler_solution, label="Numerische Lsung (eEuler)", linestyle="--", color="red")
plt.plot(t_list, iEuler_solution, label="Numerische Lsung (iEuler)", linestyle="--", color="green")
plt.plot(t_list, rk4_solution, label="Numerische Lsung (RK4)", linestyle="--", color="yellow")


plt.xlabel("t")
plt.ylabel("u(t)")
plt.title("Vergleich der analytischen und numerischen Lsungen (Aufgabe 4a)")
plt.legend()
plt.grid()
# plt.show()
plt.savefig("Aufgabe_4a.png", dpi=300)



```
